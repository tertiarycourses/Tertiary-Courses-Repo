# Principal Component Analysis(PCA)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
X=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
X.shape
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_std = scaler.fit_transform(X)
X_std
```

```python
cov_mat= np.cov(X_std.T)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
```

```python
tot = sum(eig_vals)
var_exp = [(i / tot) for i in sorted(eig_vals, reverse=True)]
```

```python
plt.figure(figsize=(6, 4))
plt.bar(range(len(eig_vals)), var_exp, alpha=0.5, align='center',label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit_transform(X_std)
```

```python
var_exp_pca= pca.explained_variance_ratio_

plt.figure(figsize=(6, 4))
plt.bar(range(len(var_exp_pca)), var_exp_pca, alpha=0.5, align='center',label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
```

```python
X_pca = pca.transform(X_std)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=cancer['target'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
